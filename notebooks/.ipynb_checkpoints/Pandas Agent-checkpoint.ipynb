```python
def print_retrieved_results(results: list):
    i = 1
    for h in results:
        print("rank: ", i)
        p = h.payload or {}
        print(f"score={h.score:.4f}  doc_type={p.get('doc_type')}  doc_id={p.get('doc_id')}")
        print("section:", p.get("section_title"))
        print("section:", p.get("table_index"))
        print("content:", (p.get("content") or ""), "...\n")
        i += 1
    return
```

```python
user_query = "What was Apple’s total debt (short-term plus long-term) at year-end 2024?"
line_items = accounting_terms_file_to_llm_digest("../data/config/SEC_accounting_terms.json")
expanded_queries = expand_query_with_ollama(
    user_query,
    allowed_line_items=line_items,
    model="qwen3:4b-instruct",
    include_original=True,
    dedupe=True,
)
queries = expanded_queries
print(queries)

client = QdrantClient(host="localhost", port=6333)
```

```python
bge = get_bge_m3_model(model_name="BAAI/bge-m3", use_fp16=False)
fused, hits_by_query = multi_query_hybrid_search_bge_m3(
    queries,
    client=client,
    collection_name="sec_docs_hybrid",
    top_k=20,
    ticker="AAPL",
    fiscal_year=2024,
    form_type="10-K",
    doc_types=["table", "table_row"],
    bge_model=bge,
    fuse=True,
    rrf_k=60,
)
deduped = dedupe_scored_points(fused, key_fn=doc_id_table_key)
enriched_cands = enrich_candidates_with_table_summaries(
    deduped,
    client=client,
    collection_name="sec_docs_hybrid",
)
query_for_reranking = queries[0] + " (" + ", ".join(queries[1:]) + ")"
print(query_for_reranking)
model = get_bge_reranker_large_model(model_name="BAAI/bge-reranker-v2-m3")
reranked = rerank_with_bge_reranker_large(
    query_for_reranking,
    enriched_cands,
    top_k=5,
    model=model,
    max_passage_chars=4000,
)
print_retrieved_results(reranked)
```

```python
def build_user_query_for_tool(user_query, df_markdown):
    return f'''
    User Query: 
    {user_query}

    Retrieved Tables:
    {df_markdown}
    '''
```

```python
%%time
from ollama import chat
import simpleeval
import json
import re
from typing import Dict, Any, Union

model_id = "llama3.1:8b"
# "qwen2.5-coder:7b"
# "deepseek-r1:14b"
# "qwen3:4b-instruct"
# "ministral-3:3b"

def _to_number(x) -> float:
    # Helper to clean "$1,200" or "(500)" to floats
    s = str(x).strip().replace("$", "").replace(",", "")
    if re.fullmatch(r"\(\s*\d+(\.\d+)?\s*\)", s):
        s = "-" + s[1:-1]
    return float(s)

# 2. STRICT TYPE HINTING (Dict[str, str] is safer than Any or String)
def financial_evaluator(variables: Dict[str, str], expression: str) -> float:
    """
    Safely evaluates a mathematical expression using provided variables.
    variables: Dict mapping names to values (e.g. {'rev': '1000'})
    """
    # Clean the dictionary values
    cleaned_vars = {k: _to_number(v) for k, v in variables.items()}
    return float(simpleeval.simple_eval(expression, names=cleaned_vars))



messages = [
    {"role": "system", "content": SYSTEM_PROMPT},
    {"role": "user", "content": build_user_query_for_tool(user_query, df_md)}
]

resp = chat(model=model_id, messages=messages
            , tools=[financial_evaluator]
            , options={"temperature": 0.0, 'num_predict': 1024})
messages.append(resp.message)

if resp.message.tool_calls:
    call = resp.message.tool_calls[0]
    result = financial_evaluator(**call.function.arguments)

    messages.append({"role": "tool", "tool_name": call.function.name, "content": str(result)})
    final = chat(model=model_id, messages=messages, tools=[financial_evaluator])
    print(final.message.content)
else:
    print(resp.message.content)
```

## Pydantic AI

```python
import re
import simpleeval
from typing import Dict

from pydantic_ai import Agent
from pydantic_ai.models.openai import OpenAIChatModel
from pydantic_ai.providers.ollama import OllamaProvider
from pydantic_ai.settings import ModelSettings
from pydantic_ai.profiles.openai import OpenAIModelProfile


SYSTEM_PROMPT = """
You are a senior financial analyst.

You must answer using ONLY the provided financial table context.

Tool available:
- financial_evaluator(variables, expression)

CRITICAL RULE:
- If the question requires ANY arithmetic (sum/difference/ratio/etc.), you MUST call financial_evaluator.
- Do NOT compute arithmetic yourself.
- After the tool returns, use the result of the tool to answer the input query.

IMPORTANT: After the tool returns a number, you MUST write a complete sentence answering the user's question with that number.
""".strip()


# ---- Model: Ollama via OpenAI-compatible endpoint ----
# (Ollama's OpenAI endpoint is typically http://localhost:11434/v1)
model = OpenAIChatModel(
    model_name="llama3.1:8b",
    provider=OllamaProvider(base_url="http://localhost:11434/v1"),
    profile=OpenAIModelProfile(
        openai_supports_strict_tool_definition=False
    ),
)

agent = Agent(
    model=model,
    system_prompt=SYSTEM_PROMPT,
    output_type=str,
    model_settings=ModelSettings(
        temperature=0.0,
        max_tokens=1024,
        extra_body={"tool_choice": "required"},
        # If you want Ollama-specific knobs (like num_predict), you can pass extra_body:
        # extra_body={"num_predict": 1024},
    ),
)


@agent.tool_plain(name="financial_evaluator")
def financial_evaluator(variables: Dict[str, str], expression: str) -> float:
    """
    Safely evaluates a mathematical expression using provided variables.
    variables: Dict mapping names -> values (strings), e.g. {"rev": "1000"}
    expression: e.g. "rev - cost"
    """
    cleaned_vars = {k: _to_number(v) for k, v in variables.items()}
    return float(simpleeval.simple_eval(expression, names=cleaned_vars))


# ---- Run it ----
prompt = build_user_query_for_tool(user_query, df_md)
result = await agent.run(prompt)

print(result.output)
```

## LangGraph

```python
from typing import Dict
import re
import simpleeval

from langchain_ollama import ChatOllama
from langchain_core.tools import tool
from langchain_core.messages import SystemMessage, HumanMessage

from langgraph.graph import StateGraph, END
from langgraph.prebuilt import ToolNode, create_react_agent
from langgraph.graph.message import add_messages
from typing_extensions import TypedDict, Annotated
```

```python
%%time
@tool
def financial_evaluator(variables: Dict[str, str], expression: str) -> float:
    """
    Safely evaluates a mathematical expression using provided variables.
    
    Args:
        variables: Dictionary mapping variable names to their string values.
                   Example: {"revenue": "1200", "cost": "450"}
        expression: Math expression using those variables. Example: "revenue - cost"
    
    Returns:
        The computed result as a float.
    """
    cleaned_vars = {k: _to_number(v) for k, v in variables.items()}
    return float(simpleeval.simple_eval(expression, names=cleaned_vars))
    
# ─────────────────────────────────────────────────────────────
# 2. AGENT SETUP
# ─────────────────────────────────────────────────────────────
SYSTEM_PROMPT = """You are a senior financial analyst.
You must answer using ONLY the provided financial table context.

CRITICAL RULES:
- If the question requires ANY arithmetic (sum/difference/ratio/etc.), you MUST call financial_evaluator.
- Extract the exact values from the table and pass them as variables.
- After receiving the tool result, provide a clear final answer to the user.

IMPORTANT: After the tool returns a number, you MUST write a complete sentence answering the user's question with that number.
"""

llm = ChatOllama(
    model="qwen3:14b",
    temperature=0.0,
    num_predict=1024,
)

agent = create_react_agent(
    model=llm,
    tools=[financial_evaluator],
    prompt=SystemMessage(content=SYSTEM_PROMPT),
)

# ─────────────────────────────────────────────────────────────
# 3. HELPER TO RUN QUERIES
# ─────────────────────────────────────────────────────────────
def run_financial_query(user_query: str, df_markdown: str) -> str:
    """Run a financial query against the agent."""
    
    full_query = f"""User Query: {user_query}

Retrieved Tables:
{df_markdown}"""
    
    result = agent.invoke({
        "messages": [HumanMessage(content=full_query)]
    })
    
    # Return the final AI message
    return result["messages"][-1].content

response = run_financial_query(user_query, df_md)
print(response)
```

## MCP

```python
import asyncio
from langchain_mcp_adapters.client import MultiServerMCPClient

from langgraph.prebuilt import create_react_agent
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_ollama import ChatOllama

from pprint import pprint

SYSTEM_PROMPT = """You are a senior financial analyst.
You must answer using ONLY the provided financial table context.

CRITICAL RULES:
- If the question requires ANY arithmetic (sum/difference/ratio/etc.), you MUST call financial_evaluator.
- Extract the exact values from the table and pass them as variables.
- After receiving the tool result, provide a clear final answer to the user.

IMPORTANT: After the tool returns a number, you MUST write a complete sentence answering the user's question with that number.
"""

llm = ChatOllama(
    model="qwen3:14b",
    temperature=0.0,
    num_predict=1024,
)

async def build_agent():
    client = MultiServerMCPClient(
        {
            "fin_math": {
                "transport": "stdio",
                "command": "python",
                "args": ["../src/rag10kq/financial_evaluator_mcp.py"],
            }
        }
    )
    tools = await client.get_tools()  # <-- MCP tools become LangChain tools
    print("Loaded tools:", [t.name for t in tools])

    agent = create_react_agent(
        model=llm,
        tools=tools,
        prompt=SystemMessage(content=SYSTEM_PROMPT),
    )
    return agent

agent = await build_agent()

async def run_financial_query(user_query: str, df_markdown: str) -> str:
    full_query = f"""User Query: {user_query}

Retrieved Tables:
{df_markdown}"""

    # result = await agent.ainvoke({"messages": [HumanMessage(content=full_query)]})
    async for event in agent.astream(
        {"messages": [HumanMessage(content=full_query)]},
        stream_mode="debug",
    ):
        pprint(event)
    # return result["messages"][-1].content
```

```python
await run_financial_query(
    user_query=user_query,
    df_markdown=df_md,   # or whatever your table markdown variable is
)
# print(response)
```
